```python
import plotly.plotly as py
import plotly.graph_objs as go 
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True) 
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
locationsessions = pd.read_csv('csvs/Location/sessions with code.csv')
locationsessions.head()
```

```python
data = dict(
        type = 'choropleth',
        locations = locationsessions['CODE'],
        z = locationsessions['Sessions'],
        text = locationsessions['Country'],
        colorscale = 'Bluered'
        
      ) 
```

```python
layout = dict(
    title = 'Total traffic',
    geo = dict(
        showframe = False,
        projection = {'type':'mercator'}
    )
)
```

```python
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap)
```
